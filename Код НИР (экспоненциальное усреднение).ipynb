```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
def get_signal_length(signal):
      return len(signal)
```

```python
file='NIR DATA.xlsx'
x1=pd.ExcelFile(file)
print(x1.sheet_names)
```

```python
df=x1.parse('Лист1')
```

```python
df
```

```python
a1 = 0.03125
df1 = [df['RHEO1'][0]*a1]
for i in range(1,get_signal_length(df)):
    df1.append(a1*df['RHEO1'][i]+(1-a1)*df1[i-1])
```

```python
plt.plot(df['RHEO1'],color='red',label='RAW')
plt.plot(df1,color='blue',label='Filtered')
plt.title("Первый канал реограммы")
plt.xlabel("Время")
plt.ylabel("Амплитуда, Ом")
plt.legend()
plt.xlim(200,7500)
plt.ylim(25,28)
```
